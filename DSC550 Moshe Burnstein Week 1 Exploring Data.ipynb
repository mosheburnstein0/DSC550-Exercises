```python
# 1 Load csv Video Games Sales with Ratings
import pandas as pd
df_video_games = pd.read_csv('Video_Games_Sales_as_at_22_Dec_2016.csv')
```

```python
# 2 Show first 10 Rows
df_video_games.head(10)
```

```python
# 3 Show Dimensions of df
df_video_games.shape
# There are 16,719 rows and 16 columns in the dataframe.
```

```python
# Top 5 games by critic score
# To account for duplicates, must print out first 7 entries
df_video_games[['Name', 'Critic_Score']].sort_values(by='Critic_Score', axis=0, ascending=False).head(7)
```

```python
# 6 First 5 games on SNES platform...with both columns shown
df_video_games[['Name', 'Platform']][df_video_games['Platform'] == 'SNES'].head()
```

```python
# 5 Number of video games in each genre
df_video_games.Genre.value_counts()
```

```python
# 6 First 5 games on SNES platform...with both columns shown
df_video_games[['Name', 'Platform']][df_video_games['Platform'] == 'SNES'].head()
```

```python
# 6 First 5 games on SNES platform
df_video_games['Name'][df_video_games['Platform'] == 'SNES'].head()
```

```python
# 7 Five publishers with highest total global sales
df_video_games[['Publisher', 'Global_Sales']].groupby('Publisher').sum().sort_values(by='Global_Sales', ascending=False).head()
```

```python
# 8 Create new column displaying percentage of global sales from North America
na_sales_share = (df_video_games['NA_Sales'] / df_video_games['Global_Sales'] * 100).astype(str) + '%'
df_video_games['NA_Sales_Share'] = na_sales_share
df_video_games.head()
```

```python
# Insert new column right after NA_Sales for easier visualization
df_video_games.insert(6, 'NA_Sales_Share', na_sales_share, True)
df_video_games.head()
```

```python
# 9 Number of NAN values in each column
df_video_games.isnull().sum()
```

```python
# 10 Median user score of all video games
df_video_games['User_Score'] = pd.to_numeric(df_video_games['User_Score'], errors = 'coerce')
median_score = df_video_games['User_Score'].median()
```

```python
median_score
```

```python
# Express NANs with median value
import numpy as np
df_video_games['User_Score'] = df_video_games['User_Score'].replace(np.nan, median_score)
df_video_games['User_Score'].head(18)
```
