```python
chance_get_in = admission_df['Chance of Admit ']
```

```python
gre_score = admission_df['GRE Score']
```

```python
toefl_score = admission_df['TOEFL Score']
```

#  Admission Data

```python
'''The adm_data.csv is a dataset of 400 rows and 9 variables about admissions to graduate school. 
The features include GRE score(a standardized admissions test for many graduate schools), TOEFL
scores which assess English as a second language, ratings(0 to 5) for the universities, statement
of purpose and letter of recommendation(both o to 5), undergraduate GPA(0 to 10), research experience
(0 or 1), and chance of admission with a 0 to 1 range. 
I plan to explore whether higher test scores will give the aspiring student greater confidence
in being accepted. If proven to be true, one might then counsel incoming students to concentrate
on test prep. Conspicuously missing from the data is the demographics of the potential graduate students.
In many institutions, influence and wealth play a greater role than academic achievement in admission chances. '''
```

```python
fig = plt.figure(figsize = (12,10))
plt.subplot(1, 2, 1)  
plt.boxplot(gre_score)
plt.title('GRE Whisker PLOT', fontsize = 25)
plt.ylabel('Test Score', fontsize = 20)

plt.subplot(1, 2, 2)
plt.boxplot(toefl_score)
plt.title('TOEFL Whisker PLOT', fontsize = 25)
plt.ylabel('Test Score', fontsize = 20)
 

plt.tight_layout()

plt.show()
```

```python
fig = plt.figure(figsize = (10,10))
hist_TOEFL = toefl_score
plt.hist(hist_TOEFL, bins=9, color='g')
plt.title('TOEFL Histogram', fontsize=24)
plt.xlabel('TOEFL Scores', fontsize = 12)
plt.ylabel('Count of Scores', fontsize = 12)
plt.show()
hist_GRE = gre_score
fig = plt.figure(figsize = (10,10))
plt.hist(hist_GRE, bins=9, color='g')
plt.title('GRE Histogram', fontsize=24)
plt.xlabel('GRE Scores', fontsize = 12)
plt.ylabel('Count of Scores', fontsize = 12)
plt.show()
```

```python
fig = plt.figure(figsize = (12,10))
plt.title('Test Scores vs Acceptance Rate', fontsize=25)
plt.xlabel('Test Scores', fontsize=15)
plt.ylabel('Chances of Admissions', fontsize=15 )
x = toefl_score
y = chance_get_in
plt.scatter(x, y, color='green', label='TOEFL Score')

x = gre_score
y = chance_get_in
plt.scatter(x, y, color='red', label='GRE Score')
plt.legend(loc='upper center', fontsize=17)
plt.show()
```

#  What the Graphs Reveal
